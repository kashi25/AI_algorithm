```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
df = pd.read_csv("test_dataset.csv")

# Create the scatter plot
plt.scatter(df['SalePrice'], df['Id'])
plt.xlabel('SalePrice')
plt.ylabel('id')
plt.title('Scatter plot of SalePrice vs id')
plt.show()
```
